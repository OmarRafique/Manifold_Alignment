```python
import itertools
import random as rn

def random_gen(low, high):
    while True:
        yield rn.randrange(low, high)

gen = random_gen(1,np.shape(data2)[0])

items = set()

# Try to add elem to set until set length is less than 10
for x in itertools.takewhile(lambda x: len(items) < abs(np.shape(data1)[0]-np.shape(data2)[0]), gen):
    items.add(x)
index=list(items)
reduced_data_set=np.delete(data2,index,0)
np.shape(reduced_data_set)
```
